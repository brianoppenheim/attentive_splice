```
import numpy as np
import pandas as pd
```

```
loss_traces = pd.read_csv('/content/drive/My Drive/bert_loss_trace.txt', sep=' ', header=None)
loss_traces = loss_traces.drop(loss_traces.columns[1], axis=1)
loss_traces.columns = ['Sequence Number', 'Loss']
seq_num = loss_traces['Sequence Number'].to_numpy()
epochs = []
curr_seq = 0
curr_epoch = 0
for s in seq_num:
  if s < curr_seq:
    curr_epoch += 1
  curr_seq = s
  epochs.append(curr_epoch)
loss_traces['Epoch'] = epochs

epoch_frames = []
for i in range(3):
  epoch_frames.append(loss_traces[loss_traces['Epoch'] == i].set_index('Sequence Number').drop(['Epoch'], axis=1))

for ep in epoch_frames:
  ep.plot()

```
